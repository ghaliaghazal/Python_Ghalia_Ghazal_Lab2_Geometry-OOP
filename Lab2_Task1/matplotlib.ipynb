```python
import matplotlib.pyplot as plt  # Import matplotlib 
from matplotlib import patches  # Import patches for shapes
from shape import Shape  

from circle import Circle 
circle_1= Circle (0, 0, 1)
circle_2= Circle (2, -1, 1)

fig, ax = plt.subplots()  # Create a figure and axis
circle_patch1 = patches.Circle((circle_1.x, circle_1.y), radius = circle_1.radius, color='blue', fill = False)  # Create first circle patch
circle_patch2 = patches.Circle((circle_2.x, circle_2.y), radius = circle_2.radius, color='red', fill = False)  
ax.add_patch(circle_patch1)  # Add first circle patch to axis
ax.add_patch(circle_patch2) 
ax.set_xlim(-3, 5) 
ax.set_ylim(-3, 5)
ax.grid() 


```
